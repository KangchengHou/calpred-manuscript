```python
%load_ext autoreload
%autoreload 2
%load_ext lab_black

import numpy as np
import pandas as pd
import pickle
import os
import admix_prs
```

```python
with open("df_res_list.pkl", "rb") as f:
    df_res_list = pickle.load(f)
```

```python
len(df_res_list)
```

```python
df = df_res_list[0].copy()
df["PRS_LOWER"] = df["PRS_MEAN"] - df["PRS_STD"] * 1.645
df["PRS_UPPER"] = df["PRS_MEAN"] + df["PRS_STD"] * 1.645
calibrate_idx = np.random.choice(df.index, size=200, replace=False)
```

```python
df
```

```python
admix_prs.calibrate_pred(
    df,
    pheno_col="PHENO_G",
    pred_col="PRS_MEAN",
    lower_col="PRS_LOWER",
    upper_col="PRS_UPPER",
    calibrate_idx=calibrate_idx,
    method="scale",
)
```
